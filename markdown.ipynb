```python
# 1. Nossa base de dados (Lista de consumos em kWh de 5 clientes)
consumos_campinas = [50, 450, 80, 600, 120]

# 2. Definição da Meta (KPI)
meta_baixo_consumo = 100

# 3. Processamento do primeiro cliente da lista
cliente_analisado = consumos_campinas[0] 

# 4. Lógica de Decisão Comercial
print(f"Analisando consumo de: {cliente_analisado} kWh")

if cliente_analisado < meta_baixo_consumo:
    resultado = "AÇÃO: Enviar Kit LED (Baixo Consumo)"
else:
    resultado = "AÇÃO: Oferecer Upgrade de Plano (Cliente Ouro)"
    

print(resultado)
```

```python
maior_consumo = max(consumos_campinas)
menor_consumo = min(consumos_campinas)

print(f"O maior consumo registrado foi: {maior_consumo} kWh")
print(f"O menor consumo registrado foi: {menor_consumo} kWh")
```

```python
# Definindo a estratégia de comunicação
cliente_alvo = consumos_campinas[2] # O cliente de 900 kWh

if cliente_alvo > 500:
    categoria = "PREMIUM"
    canal = "Ligação Consultiva Personalizada"
    oferta = "Instalação de Painéis Fotovoltaicos"
elif cliente_alvo < 100:
    categoria = "ECONÔMICO"
    canal = "SMS / E-mail Marketing"
    oferta = "Cupom para Lâmpadas LED"
else:
    categoria = "REGULAR"
    canal = "Notificação no App"
    oferta = "Dicas de Consumo Consciente"

print(f"--- ESTRATÉGIA DE MARKETING ---")
print(f"Perfil do Cliente: {categoria}")
print(f"Onde atacar: {canal}")
print(f"O que oferecer: {oferta}")
```

```python
# --- IMPORTANDO AS BIBLIOTECAS DE GRÁFICOS ---
import matplotlib.pyplot as plt
import seaborn as sns

# --- NOSSOS DADOS (Do Desafio Anterior) ---
consumos = [50, 450, 80, 600, 120, 300, 40, 900, 150, 210]
# Criamos categorias para o gráfico (uma lista de textos correspondente)
categorias = ['Econômico', 'Ouro', 'Econômico', 'Premium', 'Padrão', 
              'Padrão', 'Econômico', 'Premium', 'Padrão', 'Padrão']

# --- CONFIGURAÇÃO VISUAL ---
# Define um tema bonito do Seaborn
sns.set_theme(style="whitegrid")

# Cria a figura (janela do gráfico) e define o tamanho
plt.figure(figsize=(10, 6))

# --- CRIANDO O GRÁFICO DE BARRAS ---
# x são os consumos, y são as categorias. 'hue' colore por categoria.
grafico = sns.barplot(x=consumos, y=categorias, hue=categorias, palette="viridis", legend=False)

# --- ADICIONANDO DETALHES PROFISSIONAIS ---
plt.title('Segmentação Comercial Neoenergia - Regional Campinas', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Consumo de Energia (kWh)', fontsize=12)
plt.ylabel('Perfil do Cliente', fontsize=12)

# Adiciona uma linha vertical na média (para dar contexto de negócio)
media = sum(consumos) / len(consumos)
plt.axvline(media, color='red', linestyle='--', label=f'Média: {media:.0f} kWh')
plt.legend() # Mostra a legenda da linha da média

# --- MOSTRAR O GRÁFICO ---
plt.tight_layout() # Ajusta o espaçamento para não cortar nada
plt.show()
```
